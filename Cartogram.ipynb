```python
eamerikka = countries[countries["CONTINENT"] == "South America"]

# filtered_df = df[df['Age'] > 30]
```

```python
eamerikka.crs
eamerikka = eamerikka.to_crs(epsg=3395)
```

```python
eamerikka['area'] = eamerikka.area / 1e6 # neliÃ¶kilometreihin
```

```python
eamerikka['density'] = eamerikka['POP_EST'] / eamerikka['area'] 
```

```python
import matplotlib.colors as mcolors
binary_cmap = mcolors.ListedColormap(["#ff42d3"])
```

```python
import mapclassify as mc



cmap = binary_cmap
background_color = '#9297C4'
text_color = '#14213d'
map_color = 'white'


#scheme = mc.NaturalBreaks(eamerikka['POP_EST'], k=5)

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection=gcrs.PlateCarree())
fig.set_facecolor(background_color)
ax.set_facecolor(background_color)

gplt.cartogram(
   eamerikka, projection=gcrs.PlateCarree(),
   scale='density', limits=(0,1), alpha=0.7, color='#310A31',
   ax=ax
)
gplt.polyplot(eamerikka, facecolor='none', edgecolor='white', linewidth=0.6, ax=ax)




# set title
ax.set_title("Cartogram of the population density\n in South America", loc='center', fontsize=15, fontweight='book', family='serif', backgroundcolor= 'none', color='#310A31')




# Save the figure
#ax.savefig(NOTEBOOK_PATH / 'house_map.png', bbox_inches='tight')
plt.savefig("populationden_sa_cartogram.png", dpi=300, bbox_inches='tight')
plt.show()
```
